```python
inputNum2 = 235283

strNum = str(inputNum2)
strList = []
for n in strNum:
    if n != '0':
        strList.append(n)
    else:
        strList.append('1')

# print(strList)

strList2 = []
for i in range(len(strList)-1):
    strList2.append(strList[i]+strList[i+1])
    
# print(strList2)

ans = 0
for pair in strList2:
    num1 = int( int(pair) / 10 )
    num2 = int(pair) % 10
    
    if num1 < num2:
        ans *= 10
        ans += num1%num2
        
    else:
        ans *= 10
        ans += int(num1/num2)

print(ans)

duple = str(ans)

dupleDic = dict()
for i in duple:
    dupleDic[i] = dupleDic.get(i, 0) + 1

ans2 = ''
for s in list(dupleDic.keys()):
    ans2 += s
    
print(ans2)
```
